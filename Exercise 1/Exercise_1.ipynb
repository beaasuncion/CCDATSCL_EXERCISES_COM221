## D. Enriching

Create at least 3 new variables:
- price_per_minimum_night
- review_rate_per_month (combine ratings + frequency)
- is_expensive (boolean flag based on price threshold)


```python
df['price_per_minimum_night'] = df['price'] / df['minimum_nights']

df['review_rate_per_month'] = df['number_of_reviews'] * df['reviews_per_month']

price_75th_percentile = df['price'].quantile(0.75)
df['is_expensive'] = (df['price'] > price_75th_percentile)

print(f"Price threshold for 'is_expensive' (75th Percentile): ${price_75th_percentile:.2f}")
print("New features created. Displaying head with new columns:")
df[['price', 'minimum_nights', 'price_per_minimum_night', 'number_of_reviews', 'reviews_per_month', 'review_rate_per_month', 'is_expensive']].head()
```

Create 1 aggregated summary table

(e.g. average price per neighborhood, number of hosts per borough)

```python
neighbourhood_summary = df.groupby('neighbourhood_group').agg(
    average_price=('price', 'mean'),
    median_price=('price', 'median'),
    total_listings=('id', 'count'),
    total_unique_hosts=('host_id', 'nunique')
).reset_index()

print("Aggregated Summary Table (Price and Host Count per Borough):")
neighbourhood_summary
```

## cap_minimum_nights_outliers

### Subtask:
Identify the 99th percentile for the 'minimum_nights' column and replace any values above this threshold with the 99th percentile value to handle extreme outliers and normalize the distribution. After this, display descriptive statistics for `minimum_nights` and the number of rows removed due to invalid prices.


**Reasoning**:
To cap the outliers in 'minimum_nights', I will first calculate its 99th percentile, which serves as the upper bound for acceptable values, and then print it for verification.



```python
percentile_99 = df['minimum_nights'].quantile(0.99)
print(f"99th percentile for minimum_nights: {percentile_99}")
```

**Reasoning**:
Now that the 99th percentile has been calculated, I will replace all 'minimum_nights' values exceeding this threshold with the 99th percentile value to cap outliers.



```python
df.loc[df['minimum_nights'] > percentile_99, 'minimum_nights'] = percentile_99
print(f"'minimum_nights' values capped at {percentile_99}.")
```

**Reasoning**:
To confirm the effect of capping and understand the distribution of the 'minimum_nights' column after outlier handling, I will display its descriptive statistics.



```python
print("Descriptive statistics for 'minimum_nights' after capping:")
print(df['minimum_nights'].describe())
```
